```python
from CryptArithmeticGenetic import CryptarithmeticSolver as CryptarithmeticSolverGen
from CryptArithmeticNorm import CryptarithmeticSolver as CryptarithmeticSolverNorm
import matplotlib.pyplot as plt
import time
```

### Evaluate a performance on 100 different cryptarithmetic puzzle with genetic algorithm and the normal bruteforce method and plot a graph

```python
PROBLEMS_5 = [
    "SEND + MORE = MONEY",
    "ZEROES + ONES = BINARY",
    "COCA + COLA = OASIS",
    "DONT + HAVE = MONEY",
    "BASE + BALL = GAMES",

]
```

```python
#for cryptarithmetic solver with genetic algorithm
time_genetic = []
for each_problem in PROBLEMS_5:
    start = time.time()
    solver = CryptarithmeticSolverGen(each_problem)
    solution = solver.solve()
    end = time.time()
    time_genetic.append(end-start)
    print(f"Problem: {each_problem}")
    solver.print_solution(solution)
    print(f"Time taken: {end-start}")
    print("\n\n")
```

```python
time_norm = []
for each_problem in PROBLEMS_5:
    print("Problem: ", each_problem)
    each_problem = each_problem.replace(" ", "")
    left, right = each_problem.split("=")
    word1, word2 = left.split("+")
    print("Word1: ", word1, len(word1))
    print("Word2: ", word2, len(word2))
    print("Right: ", right, len(right))
    solverNorm = CryptarithmeticSolverNorm(word1, word2, right)
    start = time.time()
    solution = solverNorm.solve()
    end = time.time()
    time_norm.append(end-start)
    print("Time taken: ", end-start)
    print("\n\n")
```

```python
plt.plot(PROBLEMS_5, time_genetic, label="Genetic Algorithm")
plt.plot(PROBLEMS_5, time_norm, label="Normal")
plt.xlabel("Problems")
plt.ylabel("Time taken")
plt.title("Time taken for solving cryptarithmetic problems")
plt.legend()
#rotate the x axis labels
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
